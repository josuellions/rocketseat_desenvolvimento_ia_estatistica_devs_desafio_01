```python
import matplotlib.pyplot as plt
import pandas as pd

# DicionÃ¡rio de faturamento
dict_faturamento = {
    'data_ref': [
        '2023-01-01', 
        '2020-02-01', 
        '2021-03-01', 
        '2022-04-01', 
        '2023-05-01',
        '2023-06-01', 
        '2020-07-01', 
        '2021-08-01', 
        '2022-09-01', 
        '2023-10-01',
        '2022-11-01', 
        '2023-12-01',
        ],
    'valor': [
        400000, 
        890000, 
        760000, 
        430000, 
        920000,
        340000, 
        800000, 
        500000, 
        200000, 
        900000,
        570000, 
        995000,
        ]
}

df_faturamento = pd.DataFrame.from_dict(dict_faturamento)


print("MÃ©dia de vendas:")
print( df_faturamento.valor.std() / df_faturamento.valor.mean() * 100)

# OrdenaÃ§Ã£o dados
df_faturamento = df_faturamento.sort_values('data_ref', ascending=True)

# GrÃ¡fico barras
fig, ax = plt.subplots()
ax.bar(df_faturamento['data_ref'], df_faturamento['valor'],label='Valor')

plt.title('GrÃ¡fico de barras vertical')
plt.xticks(rotation=75)
plt.xlabel('Data')
plt.ylabel('Valor')
ax.legend()

plt.tight_layout()
plt.show()
plt.close()

# GrÃ¡fico de linha
fig, ax = plt.subplots()
ax.plot(df_faturamento.data_ref, df_faturamento.valor,  marker='o', color='#1f77b4', label='Faturamento')
ax.legend()

plt.xticks(rotation=75)

plt.title('GrÃ¡fico em linhas')
plt.xlabel('Data')
plt.ylabel('Valor')

plt.tight_layout()
plt.show()
plt.close()
```
